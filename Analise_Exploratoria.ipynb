```python
# Criando o Dataframe
df = pd.read_excel('AdventureWorks.xlsx')
```

```python
# Vizualizando as primeiras 5 linhas
df.head()
```

```python
# Verificando a quantidade de linhas e colunas
df.shape
```

```python
# Verificando os tipos de dados
df.dtypes
```

```python
# Qual a receita total
round(df['Valor Venda'].sum(),2) # colocando somente 2 casas ap√≥s a ','
```

```python
# Calculo do custo total
df['custo'] = df['Custo Unit√°rio'].mul(df['Quantidade']) # criando a coluna custo
```

```python
df.head(1)
```

```python
# Qual o custo total
round(df['custo'].sum(), 2)
```

```python
# Calculando o lucro total
# Criando coluna 'lucro' = 'receita' - 'custo'
df['lucro'] = df['Valor Venda'] - df['custo']
```

```python
df.head(1)
```

```python
# Total lucro
round(df['lucro'].sum(), 2)
```

```python
# J√° que temos "data da venda" e "data do envio", vamos criar uma coluna para analisar o tempo para o envio
df['tempo_envio'] = df['Data Envio'] - df['Data Venda']
```

```python
df.head(1)
```

```python
#Transformando "temp de envio" que est√° no formato "20 days" (string) em somento n√∫mero 20.
df['tempo_envio'] = (df['Data Envio'] - df['Data Venda']).dt.days
```

```python
df.head(1)
```

```python
# Verificando o tipo de coluna 'tempo_envio'
df['tempo_envio'].dtype
```

```python
# M√©dia do tempo de envio por marca
round(df.groupby('Marca')['tempo_envio'].mean(), 2)
```

# **`Missing Values`**

```python
# Verificando se temos dados faltantes
df.isnull().sum()
```

```python
#formatando as casas decimais do lucro
pd.options.display.float_format = '{:20,.2f}'.format
```

```python
# Se quisermos saber o "lucro" por "ano" e por "marca"
df.groupby([df['Data Venda'].dt.year, 'Marca'])['lucro'].sum()
```

```python
# Resetando o index
lucro_ano = df.groupby([df['Data Venda'].dt.year, 'Marca'])['lucro'].sum().reset_index()
lucro_ano
```

```python
# Qual o total de produtos vendidos?
df.groupby('Produto')['Quantidade'].sum().sort_values(ascending=False)
```

```python
# Gr√°fico total dos produtos vendidos
df.groupby('Produto')['Quantidade'].sum().sort_values(ascending=True).plot.barh(title='Total Produtos Vendidos')
plt.xlabel('Total')
plt.ylabel('Produto');
```

```python
df.groupby(df['Data Venda'].dt.year)['lucro'].sum().plot.bar(title='Lucro X Ano')
plt.xlabel('Ano')
plt.ylabel('Receita');
```
